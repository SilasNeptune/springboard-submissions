```python
import pandas as pd
import json
from pandas.io.json import json_normalize
```

```python
#Read file as DataFrame
json.load((open('Downloads/data_wrangling_json/data/world_bank_projects.json')))
json_df = pd.read_json('Downloads/data_wrangling_json/data/world_bank_projects.json')
```

```python
#Find the 10 countries with most projects
grouped_country = json_df.groupby('countryname')['countryname'].count()
print(grouped_country.sort_values(ascending=False)[:10])
```

```python
#flatten mjtheme_namecode
df = json.load((open('Downloads/data_wrangling_json/data/world_bank_projects.json')))
json_norm = json_normalize(df, 'mjtheme_namecode', ['id', ['countryname']])
print(json_norm.head(10))
```

```python
#Find the top 10 major project themes out of 11
#using code to not run into issues with missing names, and also not including the duplicates
code_count = json_norm.groupby('code')['id'].nunique() 
print(code_count.sort_values(ascending=False)[:10]) 
```

```python
# Create a dataframe with the missing names filled in

theme_dict = {} #initialize empty dict

i = 0 
for i in range(len(json_norm)): # for every row in the json_norm dataframe
    if json_norm.loc[i, 'name'] != '': # if the name is not blank
        theme_dict[json_norm.loc[i, 'code']] = json_norm.loc[i, 'name'] # set the keys and values
        
i = 0
for i in range(len(json_norm)): # for every row in the json_norm dataframe
    if json_norm.loc[i, 'name'] == '': # if the name is blank
        json_norm.loc[i, 'name'] = theme_dict[json_norm.loc[i, 'code']]

        
print(theme_dict)
json_norm['name'].value_counts().head(11) #show that there are no missing values
```
